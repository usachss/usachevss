```python
import math
import numpy as np
import matplotlib.pyplot as plot
 
class Generator():
    source=None
    def __init__(self, ampl, ampl1):
        self.amplituda = ampl
        self.amplituda1 = ampl1
    
    def createsin(self):
        self.basetok = ((math.sin(i/180*math.pi))*self.amplituda for i in range(0,720,18))
        for i in self.basetok:
            yield i

    def createsinKZ(self):
        self.basetokKZ = ((math.sin(i/180*math.pi))*self.amplituda1  for i in range(720,3600,18))
        for i in self.basetokKZ:
            yield i


        
```
